```python
#!/usr/bin/env python3
#
# Python yield
#

def squares(n):
    """Yield squares <= n."""
    i = 0
    while i * i <= n:
        yield i * i
        i += 1

print([ s for s in squares(100) ])
```

```python
#!/usr/bin/env python3
#
# Python yield
#

def squares(n):
    """Yield squares <= n."""
    i = 0
    while i * i <= n:
        yield i * i
        i += 1
    return


for s in squares(100):
    print(s, end=', ')
print()

sl = [ s for s in squares(100) ]
print(sl)
print(sl)

sg = ( s for s in squares(100) )
print(sg)
print(list(sg))
print(list(sg))

print(type({ x for x in range(10) }))
```

```python
#!/usr/bin/env python3
#
# Perfect numbers
#
import math, time

# 28 is a perfect number.
1 + 2 + 4 + 7 + 14 # + 28
print(28, [1, 2, 4, 7, 14,])

# Calculate the first 6 perfect numbers.
t = time.time()
n = 10 ** 5
for i in range(2, n):
    # Create proper divisor generator, not including 1.
    pdg = { x for sub in ( ( d, i // d, ) for d in range(2, int(math.sqrt(i)) + 1) if i % d == 0 ) for x in sub }
    if i == sum(pdg) + 1:
        print(f"{i} {time.time() - t:.4f}")
        t = time.time()
```

```python
#!/usr/bin/env python3
#
# Primes
#

def primes():
    """Yield prime numbers."""
    n = 2
    while True:
        yield n     # STUB
        n += 1

for p in primes():
    # print(p)
    if p > 10 ** 6:
        print(p)
        break
```
