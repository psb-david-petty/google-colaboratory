# `thanksgiving.py`

This is an example of generating test data for unit testing a l&agrave; [Codingbat.com](https://codingbat.com/prob/p290566?parent=/home/david_petty@psbma.org) for the Python [`doctest`](https://docs.python.org/3/library/doctest.html) module.

```
#!/usr/bin/env python3
#
# thanksgiving.py
#
import os, sys

YEAR = 2000   # not a leapyear
NOV1 = 3      # 2000/11/01 falls on a Wednesday
THURSDAY = 4  # (Sun = 0, Mon = 1, ..., Sat = 6)


def thanksgiving(year):
    """Return day of month in November on which USA Thanksgiving falls.

>>> thanksgiving(2020)
26
>>> thanksgiving(2245)
27
>>> thanksgiving(12345)
22
>>> thanksgiving(67890)
27

"""
    years = year - YEAR
    leap_days = years // 4 - years // 100 + years // 400
    nov1_day = (NOV1 + years * 365 + leap_days) % 7
    return (THURSDAY - nov1_day + 7) % 7 + 1 + 21


def generate():
    for year in range(2000, 3000):
        print(f"{year}, {thanksgiving(year)}")
    for year in range(3000, 50001, 4700):
        print(f"{year}, {thanksgiving(year)}")
    for year in [2020, 2245, 12345, 67890, ]:
        print(f">>> thanksgiving({year})\n{thanksgiving(year)}")


if __name__ == '__main__':
    is_idle, is_pycharm, is_jupyter = (
        'idlelib' in sys.modules,
        int(os.getenv('PYCHARM', 0)),
        '__file__' not in globals()
    )
    if any((is_idle, is_pycharm, is_jupyter,)):
        generate()
        import doctest
        doctest.testmod(verbose=True)
```
